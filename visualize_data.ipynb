```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
del_sizes = [40, 80, 120, 160, 200, 240, 280]
df = pd.DataFrame({})
for size in del_sizes:
    curr_df = pd.read_csv(f"tmp_sim_data_cluster/results_del_size_{size}_v3.csv")
    curr_df.columns = ['deletion_size', 'model_type', 'sensitivity', 'precision', 'tp', 'fp', 'fn', 'tn']
    
    df = df.append(curr_df)
    
df = df.reset_index()
```

```python
df['deletion_size'] = df['deletion_size']/4
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a consistent style for the plots
sns.set(style="whitegrid")

# Create the sensitivity plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="deletion_size", y="sensitivity", hue="model_type", style="model_type", data=df, marker="o")

# Customize the sensitivity plot
plt.title("Model Sensitivity")
plt.xlabel("Deletion Size (Mb)")
plt.ylabel("Sensitivity")
plt.legend(title="Model Type", loc="lower right")

# Show the sensitivity plot
plt.show()

# Create the precision plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="deletion_size", y="precision", hue="model_type", style="model_type", data=df, marker="o")

# Customize the precision plot
plt.title("Model Precision")
plt.xlabel("Deletion Size (Mb)")
plt.ylabel("Precision")
plt.legend(title="Model Type", loc="lower right")

# Show the precision plot
plt.show()
```
